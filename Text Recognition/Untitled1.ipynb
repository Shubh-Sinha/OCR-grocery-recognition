```python
from bs4 import BeautifulSoup
import requests
import re
from time import sleep
import csv
import pandas as pd
```

```python
url = "https://www.flipkart.com/search?q=cheese"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
cards = soup.find_all('div', {"class":"_4ddWXP"})
```

```python
len(cards)
```

```python
product_list = []
for card in cards:
    data_dict = {}
    data_dict['Similar Product'] = card.find('a',{"class":"s1Q9rs"}).text.strip()
    data_dict['Price'] = card.find('div',{"class":"_30jeq3"}).text.strip()
    data_dict['Rating'] = card.find('div',{"class":"_3LWZlK"}).text.strip()
    product_list.append(data_dict)
```

```python
df = pd.DataFrame(product_list)
df = df.sort_values('Rating', axis = 0, ascending = False).reset_index()
df.drop('index', axis=1)
```
